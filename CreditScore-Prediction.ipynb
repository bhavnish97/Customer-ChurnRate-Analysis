```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("churn_data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['age'])
```

```python
df = df.drop(columns=['rewards_earned'], axis=1)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['housing'].value_counts()
```

```python
housing = pd.get_dummies(data=df['housing'],drop_first=True)
```

```python
df['payment_type'].value_counts()
```

```python
payment_type = pd.get_dummies(data=df['payment_type'],drop_first=True)
```

```python
payment_type.rename(columns = {'na': 'na_payment_type'}, inplace=True)
```

```python
payment_type
```

```python
df['zodiac_sign'].value_counts()
```

```python
zodiac_sign = pd.get_dummies(data=df['zodiac_sign'],drop_first=True)
```

```python
df = pd.concat([df, housing, payment_type, zodiac_sign], axis=1)
```

```python
df
```

```python
df = df.drop(columns = ['housing', 'zodiac_sign', 'payment_type'], axis=1)
```

```python
df
```

```python
df.info()
```

```python
null_credit_scores = df[df['credit_score'].isnull()]
```

```python
X_test = null_credit_scores.drop('credit_score', axis=1)
```

```python
X_test
```

```python
y_test = null_credit_scores['credit_score']
```

```python
credit_scores_notNull = df.dropna()
```

```python
credit_scores_notNull
```

```python
X_train = credit_scores_notNull.drop('credit_score', axis=1)
```

```python
X_train
```

```python
y_train = credit_scores_notNull['credit_score']
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_Xtrain = scaler.fit_transform(X_train)
scaled_Xtest = scaler.transform(X_test)
```

```python
scaled_Xtrain
```

```python
scaled_Xtest
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
X_test['credit_score'] = y_pred
```

```python
X_test
```

```python
credit_score = credit_scores_notNull.drop('credit_score', axis=1)
```

```python
credit_score = pd.concat([credit_score, credit_scores_notNull['credit_score']], axis=1)
```

```python
credit_score
```

```python
final_df = pd.concat([credit_score, X_test])
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.to_csv('credit_score_churn_data.csv', index=False)
```
